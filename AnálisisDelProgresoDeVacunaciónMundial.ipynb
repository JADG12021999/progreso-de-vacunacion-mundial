a. Extraer la información del archivo


```python
import pandas as pd
vacunacion=pd.read_csv('/content/sample_data/country_vaccinations.csv')
```


b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurandote que las columnas con fechas sean del tipo datetime64.


```python
vacunacion['date']=pd.to_datetime(vacunacion['date'])
vacunacion.info()
vacunacion.head()
```

c. Determinar la cantidad de vacunas aplicadas de cada compañía (en base a como lo reporta cada país en la columna vaccines)


```python
vacunacion[['vaccines','source_name']].groupby('source_name').count()


```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo

```python
print('la cantidad de vacunas en el mundo es:',vacunacion['total_vaccinations'].sum())
print('la cantidad de vacunas aplicadas en el mundo es:',vacunacion['people_vaccinated'].sum())
print('la cantidad de gente vacunada completamente en el mundo es:',vacunacion['people_fully_vaccinated'].sum())
```

e. Calcular el promedio de vacunas aplicadas por país

```python
vacunacion[['country','people_vaccinated']].groupby('country').mean()
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo

```python
vacunacion=vacunacion[vacunacion['date']=="29-01-21"]
vacunacion[['daily_vaccinations','date']].groupby('date').sum()


```

g. Crear un dataframe nuevo denoninado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw

```python

conDiferencias=pd.read_csv('/content/sample_data/country_vaccinations.csv')
conDiferencias['date']=pd.to_datetime(conDiferencias['date'])
conDiferencias['diferencias']=conDiferencias['daily_vaccinations']-conDiferencias['daily_vaccinations_raw']
conDiferencias['diferencias']
```

h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua

```python
print('Fecha del primer reporte: ', conDiferencias['date'].min())
print('Fecha del último reporte: ', conDiferencias['date'].max())
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter / )

```python
conCantidad=pd.read_csv('/content/sample_data/country_vaccinations.csv')
conCantidad['date']=pd.to_datetime(conCantidad['date'])
conCantidad['canVac']=conCantidad['vaccines'].str.split('/').str.len()
conCantidad['canVac']

```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20=pd.read_csv('/content/sample_data/country_vaccinations.csv')
antes20['date']=pd.to_datetime(antes20['date'])
antes20=antes20.query("date < '20-12-20'")
antes20.head()

```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer=pd.read_csv('/content/sample_data/country_vaccinations.csv')
pfizer['date']=pd.to_datetime(pfizer['date'])
pfizer=pfizer[pfizer['vaccines'].str.contains('Pfizer')] 
pfizer.head()

```

l. Almacenar los dataframes generados (conDiferencias,conCantidad,antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter

```python
with pd.ExcelWriter("/content/sample_data/resultadosReto.xlsx") as writer:
   
    # use to_excel function and specify the sheet_name and index
    # to store the dataframe in specified sheet
    
    conDiferencias.to_excel(writer, sheet_name='conDiferencias')
    conCantidad.to_excel(writer, sheet_name='conCantidad')
    antes20.to_excel(writer, sheet_name='antes20')
    pfizer.to_excel(writer, sheet_name='pfizer')
```
